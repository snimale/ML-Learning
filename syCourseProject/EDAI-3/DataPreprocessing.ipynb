```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import word_tokenize

sentences = ' '.join(df['tweet'].str.encode('utf-16').str.decode('utf-16'))

# Tokenize the text using nltk
nltk.download('punkt')  # Download the Punkt tokenizer data (only need to run this once)

words = word_tokenize(sentences)

# Join the tokens back to form a single string for the word cloud
processed_text = ' '.join(words)

# Generate the word cloud with the 'Arial Unicode MS' font
wordcloud = WordCloud(width=800, height=400, background_color='white', font_path='/content/drive/MyDrive/Datasets/Datasets/arial-unicode-ms.ttf').generate(processed_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

#### Removing Stop Words

```python
import csv
stop_word_set = set()
with open("/content/drive/MyDrive/Datasets/Datasets/Tweets TTV ds/listOfStopWords.txt", 'r', newline='') as csvfile:
  csv_reader = csv.reader(csvfile)
  for row in csv_reader:
    if row:
      stop_word_set.add(row[0].strip())
print(stop_word_set)
```

```python
for index, row in df.iterrows():
  curr_words=row[0].split(' ')
  curr_str=row[0]
  for i in curr_words:
    if i in stop_word_set:
      curr_str=re.sub(i, "", curr_str)
  df.iloc[index, 0]=curr_str
```

#### Removing Pronouns

```python
with open('/content/drive/MyDrive/Datasets/Datasets/listOfPronouns.txt', 'r') as csvfile:
  pronouns_csv = csvfile.read()
  pronouns_list = pronouns_csv.split(',')
  pronouns_set = set(pronouns_list)
  print(pronouns_set)
```

```python
for index, row in df.iterrows():
  curr_words=row[0].split(' ')
  curr_str=list()
  for i in curr_words:
    if i not in pronouns_set:
      curr_str.append(i)
  df.iloc[index, 0]=" ".join(curr_str)
```
